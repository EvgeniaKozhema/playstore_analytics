```python
# импортируем библиотеки

import pandas as pd
```

```python
# читаем данные 

playstore = pd.read_csv('playstore.csv')
```

```python
data_head = playstore.head(3) # первые 3 строки 
data_tail = playstore.tail(3) # последние 3 строки

n_col = playstore.shape[1] # количество колонок
n_row = playstore.shape[0] # количество строк
```

```python
# считаем количество пропущенных значений в рейтинге
rating = 0
for i in playstore['Rating'].isnull():
    if i == True:
        rating += 1
```

```python
# создание срезов
slice1 = playstore.iloc[:3]
slice2 = playstore.iloc[6:9]
slice3 = playstore.iloc[16:20]

# выбор нужных колонок
columns = ['App', 'Size', 'Genres', 'Current Ver']
df_final = pd.concat([slice1[columns], slice2[columns], slice3[columns]])


# сохранение в csv
df_final.to_csv('final_dataframe.csv')
df_final
```

```python
# датафрейм, который будет включать в себя: 
df1 = playstore.loc[:2, ['App', 'Size', 'Genres', 'Current Ver']] # первые три строки исходного датафрейма и колонки: App, Size, Genres, Current Ver
df2 = playstore.loc[5:7, ['App', 'Size', 'Genres', 'Current Ver']] # строки 6-8 (включительно) и колонки: App, Size, Genres, Current Ver
df3 = playstore.loc[15:18, ['App', 'Size', 'Genres', 'Current Ver']] # строки 16-19 (включительно) и колонки: App, Size, Genres, Current Ver
df_final = pd.concat([df1, df2, df3])

df_final.to_csv('p1.csv')
pd.read_csv('p1.csv') 
```

```python
# убираем дубликаты с колонки App и сбрасываем индекс

unique_playstore = playstore.drop_duplicates(subset = 'App').reset_index(drop = True)
unique_playstore
```

```python
# подсчет доли бесплатных приложений

# Удалить дубликаты
data_no_duplicates = playstore.drop_duplicates(subset='App')
# Посчитать количество платных и бесплатных приложений
counts = data_no_duplicates['Type'].value_counts()
# Посчитать долю бесплатных приложений
free_apps_proportion = counts['Free'] / counts.sum()
# Вывести результат с округлением до двух знаков после запятой
print(round(free_apps_proportion, 2))
```

```python
# преобразование названий колонок в нижний регистр и замена пробела на нижнеее подчеркивание 
playstore.columns = playstore.columns.str.lower().str.replace(' ', '_')

playstore = playstore.drop_duplicates(subset = 'app').reset_index(drop = True) # удаление дупликатов 
```

```python
# отбор образовательных приложений с количеством отзывов более 1000 
education_playstore = playstore[ ( playstore['category'] == 'EDUCATION' )  & ( playstore ['reviews'] > 1000) ].reset_index(drop = True)
```

```python
# сводная таблица по данным
pivot_table_df = pd.pivot_table(playstore, index=['category', 'type'], values=['price', 'rating', 'reviews'], aggfunc='mean')
pivot_table_df = pivot_table_df.round({'price': 2, 'rating': 1, 'reviews': 2})

pivot_table_df.rename(columns={'price': 'mean_price', 'rating': 'mean_rating', 'reviews': 'mean_reviews'}, inplace=True)

pivot_table_df.to_csv('result.csv', sep=',')
```
